```python
def read_file(filename): #пока пишется только для файлов pos
    f =  open(filename,'r')
    mass = []
    for i in range(9): #пропускаем шапку
        f.readline() 
    for line in f:
        a = line.split()
        mass.append(tuple(map(int,a[:2]))+tuple(map(float,a[2:5])))
    f.close()
    return mass
```

```python
def carbon(graph):
    tmp = {i:0 for i in range(3,13)}
    cycles = nx.minimum_cycle_basis(graph)
    for cycle in cycles:
        tmp[len(cycle)] += 1
    return tmp
        
```

```python
def neigh(Graph, Test):
    tmp = []
    for elem in Graph.nodes():
        tmp.append(list(map(lambda y:Test.has_edge(elem,y),nx.neighbors(Graph,elem))))
    return tmp
```
